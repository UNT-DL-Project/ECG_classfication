# Baseline Models 

Logistic regressiong code and results for Training, testing and shuffled dataset. We prepaered shuffled dataset by shuffling only the label columns but retaining the rest of the columns same.

```r
data_log_reg <- input_data_rm_cor  

# Reshuffling all row in the dataset
data_log_reg <- data_log_reg[sample(1:nrow(data_log_reg)), ]

data_log_reg$affected <- as.factor(data_log_reg$affected)
levels(data_log_reg$affected) <- c("Normal", "Disease_1", "Disease_2", "Disease_3", "Disease_4")

# Reshuffling the the affected row in the dataset
data_log_shuf <- transform(data_log_reg, affected = sample(affected))

# We are dividing data into 70% training set and 30% test set
intrain <- caret::createDataPartition(y = as.factor(data_log_reg$affected), p = 0.7, list = FALSE)
intrain_shuf <- caret::createDataPartition(y = as.factor(data_log_shuf$affected), p = 0.7, list = FALSE)

#creating training data set by selecting the output row values
training <- data_log_reg[intrain,]
training_shuf <- data_log_shuf[intrain_shuf,]

#creating test data set by not selecting the output row values
testing <- data_log_reg[-intrain,]
testing_shuf <- data_log_shuf[-intrain_shuf,]

# Z-Normalization of data
training[,-ncol(training)] <- normalize(training[,-ncol(training)])
training_shuf[,-ncol(training_shuf)] <- normalize(training_shuf[,-ncol(training_shuf)])
testing[,-ncol(testing)] <- normalize(testing[,-ncol(testing)])
testing_shuf[,-ncol(testing_shuf)] <- normalize(testing_shuf[,-ncol(testing_shuf)])

l1_grid = 2^seq(10,-10, length = 3)

## fit a logistic regression model with the training dataset
L1_reg.model <- caret::train(affected ~.,
                             data = training,
                             trControl = trainControl(method = "repeatedcv", 
                                                      number = 5, 
                                                      classProbs=TRUE),
                             metric =  "Accuracy",
                             method = "glmnet",
                             lambda = l1_grid,
                             family = "multinomial", 
                             type.multinomial = "grouped"
)

## to predict using logistic regression model, probabilities obtained
L1_reg.predictions <- stats::predict(L1_reg.model, 
                                     newdata = testing, 
                                     lambda = L1_reg.model$bestTune$lambda,
                                     alpha = L1_reg.model$bestTune$alpha,
                                     type = "raw")

# Calculation of Accuracy of the model
con_matrix_lr_reg <- caret::confusionMatrix(L1_reg.predictions, testing$affected)

accuracy_lr_reg_test <- unname(con_matrix_lr_reg$overall["Accuracy"])

# Training data prediction
L1_reg.predictions_train <- predict(L1_reg.model, 
                                    newdata = training,
                                    lambda = L1_reg.model$bestTune$lambda,
                                    alpha = L1_reg.model$bestTune$alpha,
                                    type = "raw")

# Calculation of Accuracy of the model
con_matrix_lr_reg_train <- caret::confusionMatrix(L1_reg.predictions_train, training$affected)

accuracy_lr_reg_train <- unname(con_matrix_lr_reg_train$overall["Accuracy"])

# Shuffled data prediction
L1_reg_shuf.model <- caret::train(affected ~., 
                                  data = training_shuf,
                                  trControl = trainControl(method = "repeatedcv", 
                                                           number = 10, 
                                                           repeats = 3),
                                  metric = "Accuracy",
                                  method = "glmnet",
                                  lambda = l1_grid,
                                  family = "multinomial", 
                                  type.multinomial = "grouped")

## to predict using logistic regression model, probabilities obtained
L1_reg_shuf.predictions <- predict(L1_reg_shuf.model, 
                                   newdata = testing_shuf, 
                                   lambda = L1_reg_shuf.model$bestTune$lambda,
                                   alpha = L1_reg.model$bestTune$alpha,
                                   type = "raw")

# Calculation of Accuracy of the model
con_matrix_lr_reg_shuf <- caret::confusionMatrix(L1_reg_shuf.predictions, testing_shuf$affected)

accuracy_lr_reg_shuf <- unname(con_matrix_lr_reg_shuf$overall["Accuracy"])

results_lr <- cbind(accuracy_lr_reg_train,
                 accuracy_lr_reg_test,
                 accuracy_lr_reg_shuf
                 )
```

```r
con_matrix_lr_reg
results_lr
```

SVM model code and results for Training, testing and shuffled dataset. 

```r
data_svm <- input_data_rm_cor 

# Reshuffling all row in the dataset
data_svm <- data_svm[sample(1:nrow(data_svm)), ]

data_svm$affected <- as.factor(data_svm$affected)
levels(data_svm$affected) <- c("Normal", "Disease_1", "Disease_2", "Disease_3", "Disease_4")

# Reshuffling the the affected row in the dataset
data_shuf <- transform(data_svm, affected = sample(affected))

# Splitting the dataset into the Training set and Test set
intrain <- caret::createDataPartition(y = data_svm$affected, p= 0.7, list = FALSE)
training <- data_svm[intrain,]
testing <- data_svm[-intrain,]

# Splitting the shuffled dataset into the Training set and Test set
intrain_shuf <- caret::createDataPartition(y = data_shuf$affected, p= 0.7, list = FALSE)
training_shuf <- data_shuf[intrain_shuf,]
testing_shuf <- data_shuf[-intrain_shuf,]

# Z-Normalization of data
training[,-ncol(training)] <- normalize(training[,-ncol(training)])
testing[,-ncol(testing)] <- normalize(testing[,-ncol(testing)])
training_shuf[,-ncol(training_shuf)] <- normalize(training_shuf[,-ncol(training_shuf)])
testing_shuf[,-ncol(testing_shuf)] <- normalize(testing_shuf[,-ncol(testing_shuf)])

## fit a logistic regression model with the training dataset
svm_model <- caret::train(affected ~., 
                          data = training,
                          trControl = trainControl(method = "repeatedcv", 
                                                    number = 5),
                          preProc = c("center", "scale"),
                          metric = "Accuracy",
                          method = "svmRadial",
                          tuneGrid = expand.grid(.sigma = c(0.01),
                                                  .C = c(0.5)))

## to predict using SVM model, classification obtained
svm.predictions <- predict(svm_model, 
                            newdata = testing,
                            sigma = svm_model$bestTune$sigma,
                            C = svm_model$bestTune$C,
                            type = "raw")

# # Making the Confusion Matrix
con_matrix_svm <- caret::confusionMatrix(svm.predictions, as.factor(testing$affected))

# Calculation of Accuracy of the model
accuracy_svm_test <- unname(con_matrix_svm$overall["Accuracy"])


# Making prediction for training data
svm.predictions_train <- predict(svm_model, 
                                  newdata = training,
                                  sigma = svm_model$bestTune$sigma,
                                  C = svm_model$bestTune$C,
                                  type = "raw")

con_matrix_svm_train <- caret::confusionMatrix(svm.predictions_train, as.factor(training$affected))

# Calculation of Accuracy of the model
accuracy_svm_train <- unname(con_matrix_svm_train$overall["Accuracy"])


# Making prediction for shuffled data
## fit a SVM model with the training dataset
svm_model_shuf <- caret::train(affected ~., 
                                data = training_shuf,
                                trControl = trainControl(method = "repeatedcv", 
                                                        number = 5),
                                preProc = c("center", "scale"),
                                metric = "Accuracy",
                                method = "svmRadial",
                                tuneGrid = expand.grid(.sigma = c(0.01, 0.1),
                                                      .C = c(0.5, 0.7)))

## to predict using SVM model, classification obtained
svm.predictions_shuf <- predict(svm_model_shuf, 
                                newdata = testing_shuf,
                                sigma = svm_model_shuf$bestTune$sigma,
                                C = svm_model_shuf$bestTune$C,
                                type = "raw")

# # Making the Confusion Matrix
con_matrix_svm_shuf <- caret::confusionMatrix(svm.predictions_shuf, as.factor(testing_shuf$affected))

# Calculation of Accuracy of the model
accuracy_svm_shuf <- unname(con_matrix_svm_shuf$overall["Accuracy"])

results_svm <- cbind(accuracy_svm_train,
                  accuracy_svm_test,
                  accuracy_svm_shuf
                  )
```

Random Forest code and results for Training, testing and shuffled dataset. 

```r
data_rf <-input_data_rm_cor
data_rf <- data_rf[sample(1:nrow(data_rf)), ]

data_rf$affected <- as.factor(data_rf$affected)
levels(data_rf$affected) <- c("Normal", "Disease_1", "Disease_2", "Disease_3", "Disease_4")

# Reshuffling the the affected row in the dataset
data_rf_shuf <- transform(data_rf, affected = sample(affected))

# Splitting the dataset into the Training set and Test set
intrain <- caret::createDataPartition(y = data_rf$affected, p= 0.7, list = FALSE)
training <- data_rf[intrain,]
testing <- data_rf[-intrain,]

# Splitting the shuffled dataset into the Training set and Test set
intrain_shuf <- caret::createDataPartition(y = data_rf_shuf$affected, p= 0.7, list = FALSE)
training_shuf <- data_rf_shuf[intrain_shuf,]
testing_shuf <- data_rf_shuf[-intrain_shuf,]

# Z-Normalization of data
training[,-ncol(training)] <- normalize(training[,-ncol(training)])
training_shuf[,-ncol(training_shuf)] <- normalize(training_shuf[,-ncol(training_shuf)])
testing[,-ncol(testing)] <- normalize(testing[,-ncol(testing)])
testing_shuf[,-ncol(testing_shuf)] <- normalize(testing_shuf[,-ncol(testing_shuf)])

## fit a logistic regression model with the training dataset
rf_model <- caret::train(affected ~.,
                          training,
                          trControl = trainControl(method = "repeatedcv" 
                                                  ,number = 5
                          ),
                          preProc = c("center", "scale"),
                          metric = "Accuracy",
                          method = "ranger",
                          tuneGrid = expand.grid(.mtry = c(5,8)
                                                , .min.node.size = c(3,5)
                                                , .splitrule = c("extratrees")
                          ))


## to predict using RF model, classification obtained
rf.predictions <- predict(rf_model, 
                          newdata = testing,
                          mtry = rf_model$bestTune$mtry,
                          min.node.size = rf_model$bestTune$min.node.size,
                          splitrule = rf_model$bestTune$splitrule,
                          type = "raw")

# # Making the Confusion Matrix
con_matrix_rf <- caret::confusionMatrix(rf.predictions, as.factor(testing$affected))


# Calculation of Accuracy of the model
accuracy_rf_test <- unname(con_matrix_rf$overall["Accuracy"])

# Creating random forest for training data
pred_train <-   predict(rf_model, 
                        newdata = training,
                        mtry = rf_model$bestTune$mtry,
                        min.node.size = rf_model$bestTune$min.node.size,
                        splitrule = rf_model$bestTune$splitrule,
                        type = "raw")

# Creating confusion matrix
con_matrix_rf_train <- caret::confusionMatrix(pred_train, as.factor(training$affected))

# Calculation of Accuracy of the model
accuracy_rf_train <- unname(con_matrix_rf_train$overall["Accuracy"])


# Creating model for shuffled data
## fit a logistic regression model with the training dataset
rf_model_shuf <- caret::train(affected ~.,
                              training_shuf,
                              trControl = trainControl(method = "repeatedcv" 
                                                        ,number = 5
                              ),
                              preProc = c("center", "scale"),
                              metric = "Accuracy",
                              method = "ranger",
                              tuneGrid = expand.grid(.mtry = c(5,8)
                                                      , .min.node.size = c(3,5)
                                                      , .splitrule = c("extratrees")
                              ))


## to predict using RF model, classification obtained
rf.predictions_shuf <- predict(rf_model_shuf, 
                                newdata = testing_shuf,
                                mtry = rf_model_shuf$bestTune$mtry,
                                min.node.size = rf_model_shuf$bestTune$min.node.size,
                                splitrule = rf_model_shuf$bestTune$splitrule,
                                type = "raw")

# Calculation of Accuracy of the model
con_matrix_rf_shuf <- caret::confusionMatrix(rf.predictions_shuf, as.factor(testing_shuf$affected))

# Calculation of Accuracy of the model
accuracy_rf_shuf <- unname(con_matrix_rf_shuf$overall["Accuracy"])


results_rf <- cbind(accuracy_rf_train,
                  accuracy_rf_test,
                  accuracy_rf_shuf
                  )
```

# Proposed model

Neural Network code and results for Training, testing and shuffled dataset. 

```r
data_nn <- data_rf <-input_data_rm_cor
# Reshuffling all row in the dataset
data_nn <- data_nn[sample(1:nrow(data_nn)), ]

data_nn$affected <- as.factor(data_nn$affected)
levels(data_nn$affected) <- c("Normal", "Disease_1", "Disease_2", "Disease_3", "Disease_4")

# Reshuffling the the affected row in the dataset
data_shuf <- transform(data_nn, affected = sample(affected))

# Splitting the dataset into the Training set and Test set
intrain <- caret::createDataPartition(y = data_nn$affected, p= 0.7, list = FALSE)
training <- data_nn[intrain,]
testing <- data_nn[-intrain,]

# Splitting the shuffled dataset into the Training set and Test set
intrain_shuf <- caret::createDataPartition(y = data_shuf$affected, p= 0.7, list = FALSE)
training_shuf <- data_shuf[intrain_shuf,]
testing_shuf <- data_shuf[-intrain_shuf,]

# Z-Normalization of data
training[,-ncol(training)] <- normalize(training[,-ncol(training)])
testing[,-ncol(testing)] <- normalize(testing[,-ncol(testing)])
training_shuf[,-ncol(training_shuf)] <- normalize(training_shuf[,-ncol(training_shuf)])
testing_shuf[,-ncol(testing_shuf)] <- normalize(testing_shuf[,-ncol(testing_shuf)])


## Create train control
trCon = trainControl(method = "cv", 
                      number = 10)

## Creating tuning grid
tune_grid_neural = expand.grid(.size = c(1, 2, 3),
                                .decay = c(10^-1, 10^-2, 10^-3))

## fit a neural network model with the training dataset
nn_model <- caret::train(affected ~.,
                          data = training,
                          trControl = trCon,
                          method = "nnet",
                          tuneGrid = tune_grid_neural,
                          metric = "Accuracy",
                          activation = "sigmoid"
)


## to predict using neural network model, probabilities obtained
nn.predictions <- predict(nn_model, 
                          newdata = testing, 
                          size = nn_model$bestTune$size,
                          decay = nn_model$bestTune$decay,
                          type = "raw")


# Creating confusion matrix
con_matrix_nn <- caret::confusionMatrix(as.factor(nn.predictions), as.factor(testing$affected))

# Calculation of Accuracy of the model
accuracy_nn_test <- unname(con_matrix_nn$overall["Accuracy"])


## to predict using neural network model, probabilities obtained
nn.predictions_train <- predict(nn_model, 
                                training, 
                                size = nn_model$bestTune$size,
                                decay = nn_model$bestTune$decay,
                                type = "raw")

# Creating confusion matrix
con_matrix_nn_train <- caret::confusionMatrix(as.factor(nn.predictions_train), as.factor(training$affected))

# Calculation of Accuracy of the model
accuracy_nn_train <- unname(con_matrix_nn_train$overall["Accuracy"])

# fit a neural network model with the shuffled dataset
nn_model_shuf <- caret::train(affected ~.,
                              data = training_shuf,
                              trControl = trCon,
                              method = "nnet",
                              tuneGrid = tune_grid_neural,
                              metric = "Accuracy",
                              activation = "sigmoid"
)

## to predict using neural network shuffled model, results obtained
nn.predictions_shuf <- predict(nn_model_shuf, 
                                testing_shuf, 
                                size = nn_model_shuf$bestTune$size,
                                decay = nn_model_shuf$bestTune$decay,
                                type = "raw")


# Creating confusion matrix
con_matrix_nn_shuf <- caret::confusionMatrix(as.factor(nn.predictions_shuf), as.factor(testing_shuf$affected))

# Calculation of Accuracy of the model
accuracy_nn_shuf <- unname(con_matrix_nn_shuf$overall["Accuracy"])


results_nn <- cbind(accuracy_nn_train,
                  accuracy_nn_test,
                  accuracy_nn_shuf
                  )
```
